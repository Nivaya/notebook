```python
import pandas as pd

titanic = pd.read_csv('train.csv')
titanic.head()
```

```python
titanic.describe()
```

```python
titanic['Age']=titanic['Age'].fillna(titanic['Age'].median())
titanic.describe()
```

```python
titanic['Sex'].unique()
```

```python
titanic.loc[titanic['Sex']=='male','Sex']=0
titanic.loc[titanic['Sex']=='female','Sex']=1

titanic.head()
```

```python
titanic['Embarked'].unique()
```

```python
titanic['Embarked']=titanic['Embarked'].fillna('S')
titanic.loc[titanic['Embarked']=='S','Embarked']=0
titanic.loc[titanic['Embarked']=='C','Embarked']=1
titanic.loc[titanic['Embarked']=='Q','Embarked']=2

titanic.describe()
```

```python
from sklearn.svm import SVC,LinearSVC
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt

predictors=['Pclass','Sex','Fare']

svm_clf = Pipeline((
#     ('Poly_features',PolynomialFeatures(degree=3)),
    ('Scaler',StandardScaler()),
    ("svm_clf", SVC(kernel="rbf", gamma=0.1, C=1000))
))

x_train = titanic[predictors][:700]
x_test = titanic[predictors][700:]
y_train = titanic['Survived'][:700]
y_test = titanic['Survived'][700:]

svm_clf.fit(x_train,y_train)



# alg=LogisticRegression()


# y_train_pred = cross_val_score(alg, titanic[predictors], y_train, cv=3)
```

```python
y_test_pred = svm_clf.predict(x_test)
len(y_test_pred[y_test_pred==y_test])/len(y_test)


```

```python
titanic_test = pd.read_csv('test.csv')
titanic_test['Age']=titanic_test['Age'].fillna(titanic_test['Age'].median())
titanic_test['Fare']=titanic_test['Fare'].fillna(titanic_test['Fare'].median())
titanic_test.loc[titanic_test['Sex']=='male','Sex']=0
titanic_test.loc[titanic_test['Sex']=='female','Sex']=1
titanic_test[predictors].describe()
```

```python
svm_clf.fit(titanic[predictors],titanic['Survived'])

y_test_pred = svm_clf.predict(titanic_test[predictors])
# len(titanic_test[titanic_test['Survived']==y_test_pred])/len(y_test_pred)

titanic_test_y = pd.read_csv('gender_submission.csv')
titanic_test_y.describe()


titanic_test_y['Survived']=y_test_pred

titanic_test_y.describe()
titanic_test_y.to_csv('submission.csv',index=False)


```
